```python
main_pattern = 'https://www.hockeyarena.net/en/'
```

```python
name_file = pd.read_csv('league_teams.csv')
url_pattern = 'index.php?p=public_team_info_basic.php&team_id={}'
```

```python
names = name_file['Unnamed: 0']
hrefs = name_file['id']
```

```python
team_data = {}
```

```python
def get_html(url):
    html = config.s.get(url, cookies=config.s.cookies)
    return html.text
```

```python
def get_soup(html):
    soup = BeautifulSoup(html, 'lxml')
    return soup
```

```python
def get_table_data(soup):
    
    table = soup.find_all('table')
    
    keys = table[0].find_all('td', align='left')
    values = table[0].find_all('td', align='right')
    
    resulting_list = zip([x.text.strip() for x in keys], [x.text.strip() for x in values])
    
    for key, value in resulting_list:
        if key not in team_data:
            team_data[key] = [value]
        elif type(team_data[key]) == list:
            team_data[key].append(value)
        else:
            team_data[key] = [team_data[key], value]
        
    
```

```python
def url_parser(hrefs):
    for id in hrefs:
        url = main_pattern + id
        get_table_data(get_soup(get_html(url)))
```
